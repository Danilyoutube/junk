```python
clean = mega.drop(columns=['Name', 'PassengerId', 'Ticket', 'Cabin'])
```

```python
clean
```

```python
only_full = clean.dropna().reset_index()
```

```python
only_full
```

```python
only_full['Embarked'] = only_full['Embarked'].astype('category')

print(only_full['Embarked'].cat.categories)

only_full['Embarked'] = only_full['Embarked'].cat.codes
```

```python
only_full['Sex'] = only_full['Sex'].astype('category')

print(only_full['Sex'].cat.categories)

only_full['Sex'] = only_full['Sex'].cat.codes
```

```python
new_train = only_full.sample(800)
```

```python
new_train
```

```python
only_full.shape
```

```python
new_test = only_full.drop(new_train.index)

new_test
```

```python
new_train.iloc[:, 2:].to_json('new_train_x.json', orient='split', index=False)
new_train.iloc[:, 1].to_json('new_train_y.json', orient='split', index=False)

new_test.iloc[:, 2:].to_json('new_test_x.json', orient='split', index=False)
new_test.iloc[:, 1].to_json('new_test_y.json', orient='split', index=False)
```

```python
grp = only_full.groupby(['Sex'])

grp['Survived'].mean()
```

```python
grp['Survived'].count()
```

```python
survivors = only_full.loc[only_full['Survived'] == 1]
survivors_female = only_full.loc[(only_full['Survived'] == 1) & (only_full['Sex'] == 0)]
survivors_male = only_full.loc[(only_full['Survived'] == 1) & (only_full['Sex'] == 1)]
```

```python
survivors.hist(column='Age', bins=30)
```

```python
survivors_female['Age'].plot.hist(bins=30, ylim=(0, 35), grid=True, legend=True, label='female')
survivors_male['Age'].plot.hist(bins=30, ylim=(0, 35), grid=True, legend=True, label='male')
```

```python
only_full.loc[only_full['Sex'] == 0]['Age'].plot.hist(bins=30, ylim=(0, 100), grid=True, legend=True, label='female', alpha=0.5)
only_full.loc[only_full['Sex'] == 1]['Age'].plot.hist(bins=30, ylim=(0, 100), grid=True, legend=True, label='male', alpha=0.5)
```
